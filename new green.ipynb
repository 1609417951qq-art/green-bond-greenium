```python
#清洗普通债券
bond = pd.read_excel(
    "BND_Bndinfo2.xlsx",
    header=1
)

# 删除单位行
bond = bond.iloc[1:]

print(bond.shape)
```

```python
print(bond.columns)
```

```python
bond = bond.dropna(subset=["信用等级"])

bond = bond[bond["信用等级"] != ""]
```

```python
bond = bond.drop(columns=[
    "赎回条款",
    "赎回日期",
    "赎回价格",
    "回售条款",
    "回售日期",
    "回售价格",
    "提前偿还条款",
    "跨市场说明"
], errors="ignore")
```

```python
bond = bond.drop_duplicates(subset="证券ID")
```

```python
bond = bond.rename(columns={
    "证券ID": "bond_id",
    "发行机构全称": "issuer",
    "发行起始日期": "issue_date",
    "期限": "maturity",
    "票面利率": "yield",
    "信用等级": "bond_rating",
    "实际发行量": "issue_size"
})
```

```python
bond["maturity"] = pd.to_numeric(bond["maturity"], errors="coerce")
bond["yield"] = pd.to_numeric(bond["yield"], errors="coerce")
bond["issue_size"] = pd.to_numeric(bond["issue_size"], errors="coerce")
```

```python
bond["green"] = 0
```

```python
bond = bond[[
    "bond_id",
    "issuer",
    "issue_date",
    "maturity",
    "yield",
    "bond_rating",
    "issue_size",
    "green"
]]
```
